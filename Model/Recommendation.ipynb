```python
from numpy import load
import numpy as np
import pandas as pd
import os
from numpy import save
from sklearn.metrics.pairwise import cosine_similarity
```

```python
books = pd.read_csv('books.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
ratings_with_name.head()
```

## Collaborative Filtering Based Recommendation System

```python
# user who take more than 200 rate
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200     
liked_By_user = x[x].index
filtering_rating = ratings_with_name[ratings_with_name['User-ID'].isin(liked_By_user)]
```

```python
# book that have have more than 50 rate from liked_By_user (famous book)
y = filtering_rating.groupby('Book-Title').count()['Book-Rating']>=50  
famous_book = y[y].index
Final_ratings = filtering_rating[filtering_rating['Book-Title'].isin(famous_book)]
```

```python
Final_ratings.drop_duplicates()
```

```python
pt = Final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt.fillna(0, inplace=True)
```

```python
similarity_score = cosine_similarity(pt)
```

```python
print(pt.shape)
print(similarity_score.shape)
```

```python
pt.to_csv('PivotTable.csv')
save('Re-Array.npy', similarity_score)
```

```python
def recommend(book_name):
    pt = pd.read_csv('PivotTable.csv',  index_col=0)
    similarity_score = load('Re-Array.npy')
    
    index = np.where(pt.index==book_name)
    if len(index[0]) > 0:
        index = index[0][0]
        similarity_items = sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1], reverse=True)[1:8]
    else:
        return None
    
    return " - ".join([pt.index[i[0]] for i in similarity_items ])
```

```python
recommend('1984')
```
